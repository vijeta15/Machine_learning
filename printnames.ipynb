```python
#final
import random
import matplotlib.pyplot as plt

gamelist = ['rock', 'paper', 'scissors']
comp_wins = 0
user_wins = 0
total_games = 0

num_games = int(input("Enter the number of games: "))
success_rate_user = []
success_rate_comp = []

for _ in range(num_games):
    comp = random.choice(gamelist)
    user = input("Enter either rock, paper, or scissors: ")

    total_games += 1

    print("Computer chose:", comp)

    if (comp == "rock" and user == "paper") or (comp == "paper" and user == "scissors") or (comp == "scissors" and user == "rock"):
        user_wins += 1
    elif (user == "rock" and comp == "paper") or (user == "paper" and comp == "scissors") or (user == "scissors" and comp == "rock"):
        comp_wins += 1

    success_rate_user.append(user_wins / total_games)
    success_rate_comp.append(comp_wins / total_games)

    print("User wins:", user_wins)
    print("Computer wins:", comp_wins)
    print("Ties:", total_games - (user_wins + comp_wins))
    print()

plt.plot(success_rate_user, label='User Success Rate', color='yellow')
plt.plot(success_rate_comp, label='Computer Success Rate', color='purple')
plt.xlabel('Games Played')
plt.ylabel('Success Rate')
plt.title('User vs Computer Success Rate')
plt.legend()
plt.show()
```

```python
# Program to display the Fibonacci sequence up to n-th term

nterms = int(input("How many terms? "))

# first two terms
n1, n2 = 0, 1
count = 0

# check if the number of terms is valid
if nterms <= 0:
   print("Please enter a positive integer")
# if there is only one term, return n1
elif nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)
# generate fibonacci sequence
else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)  # For reproducibility
n_samples = 10
x = np.random.rand(n_samples, 1)
y = np.random.rand(n_samples, 1)

# Scatter plot of the generated data
plt.scatter(x, y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Linear regression
SLR = LinearRegression()
SLR.fit(x, y)
pred = SLR.predict(x)
mse = mean_squared_error(y, pred)

# Plot the regression line
plt.scatter(x, y)
plt.plot(x, pred, color='red', marker='*')
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

# Print results
print("MSE:", mse)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/MUL2.csv')

# Assuming your dataset has columns 'IV1', 'IV2', and 'DV'
X = df[['IV1', 'IV2']]  # Independent variables
y = df['DV']  # Dependent variable

# Scatter plot of the generated data
plt.scatter(X['IV1'], y, label='IV1')
plt.scatter(X['IV2'], y, label='IV2')
plt.xlabel("Independent Variables")
plt.ylabel("Dependent Variable")
plt.legend()
plt.show()

# Multivariate Linear regression
MLR = LinearRegression()
MLR.fit(X, y)
pred = MLR.predict(X)
mse = mean_squared_error(y, pred)

# Plot the regression lines for each variable
plt.scatter(X['IV1'], y, label='IV1')
plt.scatter(X['IV2'], y, label='IV2')
plt.plot(X, pred, color='red', marker='*')
plt.xlabel("Independent Variables")
plt.ylabel("Dependent Variable")
plt.legend()
plt.show()

# Print results
print("MSE:", mse)
print("Intercept:", MLR.intercept_)
print("Coefficients:", MLR.coef_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/pos_sAL.csv') # read the dataset
dataset
```

```python
X = dataset.iloc[:, 1:-1].values # extracts features from the dataset
y = dataset.iloc[:, -1].values # extracts the labels from the dataset
```

```python
from sklearn.linear_model import LinearRegression # import the Linear Regression model
lin_reg = LinearRegression() # creat model object
lin_reg.fit(X, y) # fits the model to the training data
```

```python
from sklearn.preprocessing import PolynomialFeatures # importing a class for Polynomial Regression
poly_regr = PolynomialFeatures(degree = 4) # our polynomial model is of order
X_poly = poly_regr.fit_transform(X) # transforms the features to the polynomial form
lin_reg_2 = LinearRegression() # creates a linear regression object
lin_reg_2.fit(X_poly, y) # fits the linear regression object to the polynomial features
```

```python
plt.scatter(X, y, color = 'red') # plotting the training set
plt.plot(X, lin_reg.predict(X), color = 'blue') # plotting the linear regression line
plt.title('Truth or Bluff (Linear Regression)') # adding a tittle to our plot
plt.xlabel('Position Level') # adds a label to the x-axis
plt.ylabel('Salary') # adds a label to the y-axis
plt.show() # prints our plot
```

```python
plt.scatter(X, y, color = 'red') # plotting the training set
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue') # plotting the polynomial regression line
plt.title('Truth or Bluff (Polynomial Regression)') # adding a tittle to our plot
plt.xlabel('Position level') # adding a label to the x-axis
plt.ylabel('Salary') # adding a label to the y-axis
plt.show() # prints our plot
```

```python
X_grid = np.arange(min(X), max(X), 0.1) # choice of 0.1 instead of 0.01 to make the graph smoother
X_grid = X_grid.reshape((len(X_grid), 1)) # reshapes the array to be a matrix
plt.scatter(X, y, color = 'red') # plots the training set
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue') # plots a polynomial regression line
plt.title('Truth or Bluff (Polynomial Regression)') # adds tittle to the plot
plt.xlabel('Position level') # adds label to the x-axis
plt.ylabel('Salary') # adds label to the y-axis
plt.show() # prints our plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
datas = pd.read_csv('/content/drive/MyDrive/pos_sAL.csv')
datas
```

```python
X = datas.iloc[:, 1:2].values
y = datas.iloc[:, 2].values
```

```python
# Features and the target variables
X = datas.iloc[:, 1:2].values
y = datas.iloc[:, 2].values

# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin = LinearRegression()

lin.fit(X, y)
```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color='blue')

plt.plot(X, lin.predict(X), color='red')
plt.title('Linear Regression')
plt.xlabel('lev')
plt.ylabel('sal')

plt.show()
```

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y, color='blue')

plt.plot(X, lin2.predict(poly.fit_transform(X)),
		color='red')
plt.title('Polynomial Regression')
plt.xlabel('level')
plt.ylabel('salary')

plt.show()
```
